```python
#importar o Pandas
import pandas as pd

#ler o arquivo csv
df = pd.read_csv("C:\\Users\\BlueShift\\Downloads\\Python\\Vendas Online\\Vendas Online.csv")

#Chamar o CSV para ver se Funciona
df

#Verifica se não tem nenhuma tabela vazia e preenche com false e true onde a buraco. 
#Sempre usar quando tratar algo, para verificar se foi Modificado e quebrado
#df.isna()
```

```python
#1)Qual é o produto mais vendido?
#Analise a quantidade total vendida para cada produto e identifique aquele com o maior volume de vendas.

#Agrupar os dados por 'Produto' e somar as quantidades vendidas
total_vendido_por_produto = df.groupby('Produto')['Quantidade'].sum().reset_index()

#Ordenar os produtos pelo total de vendas em ordem decrescente
produto_mais_vendido = total_vendido_por_produto.sort_values(by='Quantidade', ascending=False).head(5)

#Chamar a Função
produto_mais_vendido
```

```python
#Manter o valor unitário da primeira linha encontrada para cada produto
valor_unitario_por_produto = df[['Produto', 'Valor Unitário']].drop_duplicates(subset=['Produto'])

#Agrupar os dados por Produto e somar as quantidades vendidas
total_vendido_por_produto = df.groupby('Produto')['Quantidade'].sum().reset_index()

#Fazer o merge dos dados para trazer o valor unitário ao lado da quantidade vendida
produto_mais_vendido = total_vendido_por_produto.merge(valor_unitario_por_produto, on='Produto')

#Ordenar os produtos pelo total de vendas em ordem decrescente
produto_mais_vendido = produto_mais_vendido.sort_values(by='Quantidade', ascending=False).head(5)

#Chamar a Função
produto_mais_vendido
```

```python
#2)Quais são as três regiões com maior faturamento?
#Calcule a receita total por região e liste as três com o maior faturamento.

#Calcular a receita total (Quantidade * Valor Unitário) por cada venda
df['Receita'] = df['Quantidade'] * df['Valor Unitário']

#Agrupar os dados por 'Localidade' e somar as receitas
faturamento_por_regiao = df.groupby('Localidade')['Receita'].sum().reset_index()

#Ordenar as regiões pelo faturamento total em ordem decrescente
top_3_regioes = faturamento_por_regiao.sort_values(by='Receita', ascending=False).head(3)

#chamar a Função
top_3_regioes
```

```python
#3)Existe uma relação entre categorias de produto e ticket médio?
#Agrupe os dados por categoria de produto e calcule o ticket médio (valor médio de cada pedido) 
#para verificar quais categorias geram mais receita por venda.

#Mesma base da atividade Anterior
df['Receita'] = df['Quantidade'] * df['Valor Unitário']

#Agrupar os dados por Categoria e calcular Receita total e Quantidade total, 
#.agg está sendo usando para agregar mais informações no mesmo codigo
categoria_sumario = df.groupby('Categoria').agg(Receita_Total=('Receita', 'sum'),Quantidade_Total=('Quantidade', 'sum')).reset_index()

#Calcular o ticket médio (Receita Total / Quantidade Total)
categoria_sumario['Ticket_Medio'] = categoria_sumario['Receita_Total'] / categoria_sumario['Quantidade_Total']

#Ordenar as categorias pelo ticket médio em ordem decrescente
categoria_sumario = categoria_sumario.sort_values(by='Ticket_Medio', ascending=False)

#Exibir o resultado
categoria_sumario
```

```python
#4)Como o faturamento varia ao longo do tempo?
#Crie uma análise temporal para observar como as vendas se comportam ao longo dos meses. 
#Identifique picos e possíveis sazonalidades que possam guiar estratégias de marketing.

#Criar uma coluna de Receita
df['Receita'] = df['Quantidade'] * df['Valor Unitário']
df['Data'] = pd.to_datetime(df['Data'])

#Agrupar os dados por mês e calcular a receita total
df['Mês'] = df['Data'].dt.to_period('M')
vendas_por_mes = df.groupby('Mês')['Receita'].sum().reset_index()
vendas_por_mes['Mês'] = vendas_por_mes['Mês'].dt.to_timestamp()

#Chamar a Função
vendas_por_mes
```

```python
#5)Quais produtos têm mais sucesso em cada região?
#Determine os produtos mais vendidos em cada região. 
#Isso permitirá que a GreenMart direcione campanhas regionais com base nas preferências locais.

#Fazer a junção dos dados por Localidade e Produto, somando a quantidade vendida
vendas_por_regiao_produto = df.groupby(['Localidade', 'Produto']).agg(
    Quantidade_Total=('Quantidade', 'sum')
).reset_index()

#Ordenar as vendas por Localidade e Quantidade Total
vendas_por_regiao_produto = vendas_por_regiao_produto.sort_values(['Localidade', 'Quantidade_Total'], ascending=[True, False])

#Para cada região, pegar o produto com a maior quantidade vendida
produtos_mais_vendidos_por_regiao = vendas_por_regiao_produto.drop_duplicates(subset='Localidade', keep='first')

#Chamando a Função
produtos_mais_vendidos_por_regiao
```
